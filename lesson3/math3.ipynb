3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
from pylab import rcParams

rcParams["figure.figsize"] = 5, 5
```

```python
def draw_circle(x0, y0, r):
    x = np.linspace(-0.1-x0-r, 0.1 + x0 + r, 10001)
    y1 = np.sqrt(r**2 - (x - x0)**2) + y0
    y2 = -np.sqrt(r**2 - (x - x0)**2) + y0
    plt.plot(x,y1)
    plt.plot(x,y2)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
```

```python
draw_circle(1,1,2)
```

```python
def draw_ellips(x0, y0, a, b):
    rcParams["figure.figsize"] = 5, 5*b/a
    x = np.linspace(x0 - a - 0.1, x0 + a + 0.1, 10001)
    y1 = b*np.sqrt(1 - (x - x0)**2/a**2) + y0
    y2 = -b*np.sqrt(1 - (x - x0)**2/a**2) + y0
    plt.plot(x,y1)
    plt.plot(x,y2)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
```

```python
draw_ellips(1,2,2,1)
```

```python
def draw_hyperbola(x0, y0, a, b):
    rcParams["figure.figsize"] = 5, 5*b/a
    x = np.linspace(x0 - a - 2, x0 + a + 2, 10001)
    y1 = b*np.sqrt(-1 + (x - x0)**2/a**2) + y0
    y2 = -b*np.sqrt(-1 + (x - x0)**2/a**2) + y0
    plt.plot(x,y1)
    plt.plot(x,y2)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
```

```python
draw_hyperbola(0,0,1,1)
```

4. Задание (в программе

Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1


```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (x + np.exp(x) - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2,7) 
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1-y) - 1)

x1, y1 =  fsolve(equations, (-1, 0))
              
print (x1, y1)
```

```python
x1, y1 =  fsolve(equations, (2, 4))
              
print (x1, y1)
```
